```python
import math
import pandas as pd
import numpy as np
```

```python
# load node to create adjendancy matrix
nodes = pd.read_csv('data/nodes.csv')
nodes.head()
```

```python
nodemap = {}
for index, row in nodes.iterrows():
    # print(row['lat'], row['lon'], index)
    nodemap[str(round(row['id']))] = [row['lat'], row['lon']]
print(nodemap)
```

```python
edges = pd.read_csv('data/edges.csv')
edges.head()
```

```python
aM = np.zeros((len(nodemap)+2, len(nodemap)+2))
for index, row in edges.iterrows():
    aM[row['from']][row['to']] = row['dis']
    aM[row['to']][row['from']] = row['dis']
print(aM)
```

```python
print('adjency matrix is ready')
```

```python
from algorithms import dijsktra, bellman_ford, floyd_warshall

start = 15
end = 92

dijsktra.solve_dijsktra(aM, start, end)

bellman_ford.solve_bellman_ford(aM, start, end)

floyd_warshall.solve_floyd_warshall(aM, start, end)
```
